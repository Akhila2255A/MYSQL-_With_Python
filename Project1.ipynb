# Visualization

```python

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = data['Gender']
y = data['BloodGroup']
plt.xlabel('Gender',fontsize=24)
plt.ylabel('BloodGroup', fontsize=24)
plt.bar(x,y)
```

```python
sns.set(color_codes=True)
```

```python
data.info()
```

```python
# Distplot()
sns.displot(data['BP'])
```

```python
data1=data[data.index.isin(range(7,100))]
sns.boxplot(x='BloodGroup',y='BillNumber',data=data1)
plt.show()
```

```python
Data1=data[data.index.isin(range(56,111))]
sns.barplot(x='BP',y='Weight',data=Data1)
```

```python
labels = data['OpReason'].head(5)
sizes= data['Weight'].head(5)
plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Pie chart ")
plt.show()
```

```python
data5=data[data.index.isin(range(99,125))]
sns.violinplot(x='BP',y='Weight',data = data5)
plt.xticks(rotation=20)
```

```python
plt.scatter(data.BP,data.Height)
plt.xlabel('BP')
plt.ylabel('Height')
plt.xticks(rotation=45)
plt.show()
```
