# Pie Chart

```python
labels = data['phone_name'].head(5)
sizes= data['cost'].head(5)
plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Pie chart ")
plt.show()
```

# Violinplot

```python
data5=data[data.index.isin(range(99,110))]
sns.violinplot(x='ram',y='cost',data = data5)
plt.xticks(rotation=20)
```

```python
Data1=data[data.index.isin(range(56,111))]
sns.barplot(x='manufacture_country',y='cost',data=Data1)
```

```python
plt.scatter(data.os_type,data.cost)
plt.xlabel('os_type	')
plt.ylabel('cost')
plt.xticks(rotation=60)
plt.show()
```
